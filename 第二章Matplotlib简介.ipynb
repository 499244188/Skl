```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(-np.pi,np.pi,200)
C,S = np.cos(x),np.sin(x)
```

```python
plt.plot(x,C,c='r')
plt.plot(x,S,c='b')
plt.show()
```

```python
plt.plot(x,C,color='blue',linewidth=2.0,linestyle='-')
plt.plot(x,S,color='red',linewidth=2.0,linestyle='-')

#设置坐标轴长度
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(C.min()*1.1,C.max()*1.1)

#设置坐标轴的刻度标签
plt.xticks((-np.pi,-np.pi/2,0,np.pi/2,np.pi),(r'$-\pi$',r'$-\pi/2$',r'0','r$+\pi/2$',r'$+\pi$'))
plt.yticks([-1,-0.5,0,0.5,1])

#坐标轴共有4个连线，通过设置透明颜色和隐藏上方和右边的边线
#通过 set_position() 移动左侧和下侧的边线
#通过 set_ticks_position() 设置坐标 轴的刻度线的显示位置
ax = plt.gca()                             #gca代表当前坐标轴 ‘get current axis'
ax.spines['right'].set_color('none')       #隐藏坐标轴
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')      #设置刻度显示位置
ax.spines['bottom'].set_position(('data',0)) #设置下方坐标轴位置
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))  #设置左侧坐标轴位置

plt.legend(['Cos','Sin'],loc='upper left')
t = 2*np.pi / 3
plt.plot([t,t],[0,np.cos(t)],color='blue',linewidth=1.5,linestyle='--')
plt.plot([t,t],[0,np.sin(t)],color='red',linewidth=1.5,linestyle='--')
plt.scatter([t,],[np.cos(t),],50,color='blue')
plt.scatter([t,],[np.sin(t),],50,color='red')
plt.scatter([t,],[0,],50,color='black')

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
            xy=(t,np.cos(t)),xycoords='data',
            xytext=(-90,-50),textcoords='offset points',fontsize=16,
           arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))

plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{sqrt(3)}{2}$',
            xy=(t,np.sin(t)),xycoords='data',
            xytext=(8,30),textcoords='offset points',fontsize=16,
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))


for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white',edgecolor='None',alpha=0.65))
```

### 优美的坐标轴
* 继承 matplotlib.ticker.Locatorn 来实现自定义刻度标签

```python
def tickline():
    plt.xlim(0,10),plt.ylim(-1,1),plt.yticks([])
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    
    ax.xaxis.set_minor_locator(plt.MultipleLocator(0.1))
    #设置刻度标签的文字大小
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_fontsize(16)
    ax.plot(np.arange(11),np.zeros(11))
    return ax
```

```python
locators = [
    'plt.NullLocator()',
    'plt.MultipleLocator(base=1.0)',
    'plt.FixedLocator(locs=[0,2,8,9,10])',
    'plt.IndexLocator(base=3,offset=1)',
    'plt.LinearLocator(numticks=5)',
    'plt.LogLocator(base=2,subs=[1.0])',
    'plt.MaxNLocator(nbins=3,steps=[1,3,5,7,9,10])',
    'plt.AutoLocator()',
    ]
n_locators = len(locators)

#计算图形对象大小
size = 1024,60*n_locators
dpi=72
figsize=size[0]/float(dpi),size[1]/float(dpi)
fig=plt.figure(figsize=figsize,dpi=dpi)
fig.patch.set_alpha(0)

for i,locator in enumerate(locators):
    plt.subplot(n_locators,1,i+1)
    ax=tickline()
    ax.xaxis.set_major_locator(eval(locator))
    plt.text(5,0.3,locator[4:],ha='center',size=16)
plt.subplots_adjust(bottom=.01,top=.99,left=.01,right=.99)
plt.show()
```

### 画图操作

* 使用 plt.fill_between() 函数来填充

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)

plt.subplot(1,2,1)
plt.scatter(X,Y,s=80,c=T,alpha=.5)

plt.xlim(-1.5,1.5)
plt.xticks(())
plt.ylim(-1.5,1.5)
plt.yticks(())
#plt.yticks((np.linspace(-2,2,10,endpoint=False)))

n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.subplot(1,2,2)
plt.plot(X,Y+1,color='blue',alpha=1.00)
plt.fill_between(X,1,Y+1,color='blue',alpha=.25)

plt.plot(X,Y-1,color='blue',alpha=1.00)
plt.fill_between(X,-1,Y-1,Y>0,color='yellow',alpha=1)
plt.fill_between(X,-1,Y-1,Y<0,color='red',alpha=.8)

plt.xlim(-np.pi,np.pi)
plt.xticks(())
plt.ylim((-2.5,2.5))
plt.yticks(())
```

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)

plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x,y+0.05,'%.2f' % y,  ha='center',va='bottom')

for x,y in zip(X,Y2):
    plt.text(x,-y-0.05,'%.2f' % x,ha='center',va='top')
plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(())
```

* plt.contourf() 函数可以填充等高线  
* cmap表示颜色风格  
* pltcontour() 画出等高线  
* 留意 np.meshgrid() 的用法  
* 最后使用plt.clable() 画出等高线的数值  

```python
def f(x,y):
    return(1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n=256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
plt.figure(figsize=(15,5))
plt.subplot(1,2,2)

# plt.contourf(X,Y,f(X,Y),8,alpha=.75,cmap=plt.cm.hot)
plt.contourf(X,Y,f(X,Y),8,alpha=.75,cmap='BuPu')
C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidths=.5)
plt.clabel(C,inline=1,fontsize=10)

plt.xticks(())
plt.yticks(())
```

### 方格图

* set_major_locator()
* set_minor_loactor()

```python
plt.figure(figsize=(8,4))
ax=plt.subplot()
ax.set_xlim(0,4)
ax.set_ylim(0,3)

ax.xaxis.set_major_locator(plt.MultipleLocator(1.0))
ax.xaxis.set_minor_locator(plt.MultipleLocator(0.1))

ax.yaxis.set_major_locator(plt.MultipleLocator(1.0))
ax.yaxis.set_minor_locator(plt.MultipleLocator(0.1))

ax.grid(which='major',axis='x',linewidth=0.75,linestyle='-',color='0.75')
ax.grid(which='minor',axis='x',linewidth=0.25,linestyle='-',color='0.25')

ax.grid(which='major',axis='y',linewidth=0.75,linestyle='-',color='0.75')
ax.grid(which='minor',axis='y',linewidth=0.25,linestyle='-',color='0.25')

ax.set_xticklabels([])
ax.set_yticklabels([])
```
