### 学习曲线

```python
n_dots = 200
x = np.linspace(0,1,n_dots)
y = np.sqrt(x) + 0.2* np.random.rand(n_dots) -.1
```

```python
# 因为sklearn的接口需要用到n_sample x n_feature的矩阵，
#需要转换为 200 X 1的矩阵
X = x.reshape(-1,1)
y = y.reshape(-1,1)
```

#### 在sklear里，需要用Pipeline 来构造多项式模型

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

#### polynomial_moedl() 函数生成一个多项式模型，其中degree 表示多项式的阶数

```python
def polynomial_model(degree=1):
    polynomial_features = PolynomialFeatures(degree=degree,
                                            include_bias=False)
    linear_regression = LinearRegression()
    #这是一个流水线，先增加多项式系数，然后在用线性回归算法拟合数据
    pipeline = Pipeline([('polynomial_featrures',PolynomialFeatures),
                         ('linear_regression',linear_regression)])
    return pipeline
```

在scikit-learn中，我们用自己实现学习曲线的算法，直接使用**sklearn.model_selection.learning_curve() 函数来画出学习曲线**，他会自动把训练样本数量按照预定的规则逐渐增加，然后画出不同训练样本数量时的模型准确性

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import learning_curve
```

```python
def plot_learning_curve(estimator,title,X,y,ylim=None,cv=None,
                       n_jobs=1,train_sizes=np.linspace(.1,1.0,5)):
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel('Score')
    train_sizes_abs,train_scores,test_scores = learning_curve(estimator,X,y,cv=cv,n_jobs=n_jobs,train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores,axis=1)
    train_scores_std = np.std(train_scores,axis=1)
    test_scores_mean = np.mean(test_scores,axis=1)
    test_scores_std = np.std(test_scores,axis=1)
    plt.grid()
    
    plt.fill_between(train_szies_abs,train_scores_mean-train_scores_std,
                    train_scores_mean+train_scores_std,alpha=0.1,color='r')
   
    plt.fill_between(train_szies_abs,test_scores_mean-test_scores_std,
                    test_scores_mean+test_scores_std,alpha=0.1,color='r')
    
    plt.plot(train_sizes_abs,train_scores_mean,'o-',color='r',label='Training score')
    
    plt.plot(train_sizes_abs,train_scores_mean,'o-',color='r',label = 'Cross-validation score')
    
    plt.legend(loc='best')
    return plt
    
    
    
```

```python
#为了让学习曲线更平滑，计算10次交叉验证数据集的分数
cv = ShuffleSplit(n_splits=10,test_size=0.2,random_state=0)
titles = ['Learning Curves(under Fitting)',
          'Learning Curves',
          'Learning Curves (Over Fitting)']
degrees = [1,3,10]

plt.figure(figsize=(18,4),dpi=200)
for i in range(len(degrees)):
    plt.subplot(1,3,i+1)
    plot_learning_curve(polynomial_model(degrees[i]),titles[i],X,y,ylim=(0.75,1.01),cv=cv)
    plt.show()
```

```python
plt.title("Learn")
plt.ylim(0.75,1,1.01)
plt.xlabel("Training examples")
plt.ylabel('Score')
train_sizes_abs,train_scores,test_scores = learning_curve(1,X,y,cv=cv,n_jobs=1,train_sizes=np.linspace(.1,1.0,5))
train_scores_mean = np.mean(train_scores,axis=1)
train_scores_std = np.std(train_scores,axis=1)
test_scores_mean = np.mean(test_scores,axis=1)
test_scores_std = np.std(test_scores,axis=1)
plt.grid()

plt.fill_between(train_szies_abs,train_scores_mean-train_scores_std,
                train_scores_mean+train_scores_std,alpha=0.1,color='r')

plt.fill_between(train_szies_abs,test_scores_mean-test_scores_std,
                test_scores_mean+test_scores_std,alpha=0.1,color='r')

plt.plot(train_sizes_abs,train_scores_mean,'o-',color='r',label='Training score')

plt.plot(train_sizes_abs,train_scores_mean,'o-',color='r',label = 'Cross-validation score')

plt.legend(loc='best')
```
