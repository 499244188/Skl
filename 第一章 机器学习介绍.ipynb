### 有监督学习 [Supervised learning]

通过大量已知输入和输出相配对的数据，让计算机从中学习出规律，从而能针对一个新的输入作出各种合理预测。

### 回归学习[Regression learning]

输出结果是一个具体数值，他的预测模型是一个连续函数

### 分类学习[Classfication learning]

输出结果是离散的

### 无监督学习[Unsupervised learning]

通过学习大量的无标记的数据，去分析数据本身的内在特点结构，比如聚类「Clustering」

### 机器学习应用开发的步骤：

-  特征选择

### 判断是否为质数

```python
a = np.arange(1,101)
n_max = int(np.sqrt(len(a)))
is_prime = np.ones(len(a),dtype=bool)
is_prime[0] = False
for i in range(2,n_max):
    if i in a[is_prime]:
        is_prime[(i**2-1)::i] = False  #减1是为了修复从0开始的问题
print(a[is_prime])
```
